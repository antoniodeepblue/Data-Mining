Utilizaremos algoritmos de clasificaci√≥n, predicci√≥n, regresi√≥n lineal, clustering (ejemplo k-means o k-nearest neighbor), Deep Learning (ej.: red neuronal), bayesiano, etc. y podr√° haber variantes si lo que vamos a procesar son im√°genes, sonido, texto, valores num√©ricos.

Tenemos decidido qu√© tipo de modelo queremos obtener. Por lo tanto, ahora hay que elegir un m√©todo de construcci√≥n de modelos entre la multitud de m√©todos que permiten obtener el tipo de modelo que nos interesa.

El proceso de construcci√≥n de modelos consiste en encontrar el modelo (el conocimiento) que responde mejor a las caracter√≠sticas impl√≠citas en los datos. Este tipo de problema suele conceptualizarse como un proceso de b√∫squeda. Los principales problemas de este tipo de conceptualizaci√≥n son los siguientes:

1) Disponer de una medida de calidad que califica como el mejor posible un modelo que solo es relativamente bueno con respecto a los modelos que tiene cerca de √©l, pero que no es el modelo √≥ptimo dentro de todo el espacio de soluciones posibles. Este problema se denomina problema de la obtenci√≥n de m√≠nimos locales.

2) Obtener un modelo que sea bueno teniendo en cuenta solo los datos de los que disponemos. Este problema que consiste en obtener modelos que generalizan mal y son demasiado espec√≠ficos con respecto a los datos que han visto se conoce como el problema de la sobreespecializaci√≥n (en ingl√©s, overfitting).

3) En el transcurso del tiempo es posible que el modelo obtenido se degenere porque empiezan a aparecer observaciones nuevas para las que no generaliza bien. Este es el problema del envejecimiento de modelos. Es necesario actualizarlos, ya sea desde cero o incrementalmente, dependiendo del tipo de modelo del que se trate.

Podemos dividir los m√©todos de miner√≠a de datos en lo que concierne a:

a) Conocimiento a priori. Reutilizaci√≥n del conocimiento extra√≠do por otro sistema de miner√≠a de datos con el fin de guiar el proceso de b√∫squeda. En ingl√©s, multistrategy learning.

b) Al tipo de datos.

1.   M√©todos que utilizan solo observaciones, es decir, solo informaci√≥n con
respecto a los valores que se han recogido, sin m√°s informaci√≥n a√±adida. Los m√©todos que parten de esta base se conocen como m√©todos no supervisados. Un ejemplo t√≠pico de ello son los m√©todos de agregaci√≥n. En ingl√©s, clustering.

2.   M√©todos que a√±aden informaci√≥n discriminando las observaciones. Los m√©todos que parten de esta situaci√≥n inicial se denominan m√©todos supervisados. Un ejemplo t√≠pico de ello es la clasificaci√≥n.

c) Al proceso de construcci√≥n:

1. M√©todos batch.

2. M√©todos incrementales.

3. M√©todos interactivos.

En nuestro ejemplo de clientes mayoristas est√°bamos buscando un m√©todo que parte de la base que utiliza solo observaciones, entonces es un m√©todo no supervisado. Un ejemplo t√≠pico de ello son los m√©todos de agregaci√≥n. En ingl√©s, clustering. Utilizaremos un algoritmo particional como k-means.

El algoritmo K-means es un m√©todo de agrupamiento que se utiliza para dividir un conjunto de datos en K grupos (o "clusters") basados en caracter√≠sticas similares. El proceso de entrenamiento del modelo K-means consta de los siguientes pasos:

1. Inicializaci√≥n de centroides: Se eligen K puntos aleatorios del conjunto de datos como los centroides iniciales. Un centroide es simplemente un punto que representa el centro de un grupo.

2. Asignaci√≥n de puntos a clusters: Cada punto de datos es asignado al cluster cuyo centroide es el m√°s cercano en t√©rminos de distancia euclidiana u otra m√©trica de distancia espec√≠fica.

3. Actualizaci√≥n de centroides: Una vez que todos los puntos han sido asignados a clusters, los centroides se recalculan tomando el promedio de todos los puntos asignados a ese cluster. Esto significa que el centroide se mueve hacia el centro de masas de los puntos asignados.

4. Repetici√≥n de pasos 2 y 3: Los pasos 2 y 3 se repiten hasta que se cumple un criterio de parada, como un n√∫mero m√°ximo de iteraciones o una convergencia suficiente.

5. Convergencia: El algoritmo eventualmente converge a una soluci√≥n donde los centroides ya no cambian entre iteraciones, lo que indica que se ha alcanzado un estado estable.

6. Resultado final: Una vez que el algoritmo ha convergido, se tienen los clusters finales y los centroides asociados.

Es importante tener en cuenta que el n√∫mero de clusters, K, debe ser especificado antes de iniciar el algoritmo. La elecci√≥n de K es una decisi√≥n crucial y puede requerir m√©todos como el codo (elbow method) o el an√°lisis de silueta para determinar el valor √≥ptimo.

El proceso de entrenamiento es relativamente r√°pido, pero el resultado final puede depender de la inicializaci√≥n aleatoria de los centroides. Por esta raz√≥n, a veces se ejecuta el algoritmo varias veces con diferentes inicializaciones y se selecciona el mejor resultado.

**üéì 5. Evaluaci√≥n  del modelo**

T√≠picamente, para algoritmos supervisados, el proceso de evaluaci√≥n consiste en disponer de dos conjuntos de datos procedentes del mismo conjunto inicial (y ya preparado): un conjunto de datos que se utiliza para construir el modelo y otro, para evaluarlo. En algunas ocasiones, se introduce un tercer conjunto entre estos dos: el conjunto de validaci√≥n. Utilizamos un conjunto de datos para construir el modelo; otro, para darlo por bueno (validarlo), y un tercero, para evaluarlo. (No es nuestro caso)

No se pueden comparar modelos predictivos con modelos descriptivos. Para los modelos predictivos se buscar√° determinar hasta qu√© punto son buenos haciendo predicciones, mientras que para los modelos descriptivos evaluaremos hasta qu√© punto se ajustan al dominio descrito.

Cuando ya tenemos un modelo que posee el nivel de calidad requerido y que ha sido validado mediante el proceso de evaluaci√≥n, hay que interpretarlo y extraer el significado del conocimiento que nos est√° mostrando. En nuestro caso es un modelo descriptivo en el que utilizamos el algoritmo no supervisado, por ejemplo, k-means.

Para evaluar un modelo K-means, puedes usar varias m√©tricas y t√©cnicas. Aqu√≠ te proporciono algunas de las m√°s comunes:

1. Inercia o Suma de Cuadrados Dentro del Grupo (WCSS - Within-Cluster Sum of Squares): Esta m√©trica mide la dispersi√≥n de los puntos dentro de cada cluster. Es el objetivo de optimizaci√≥n del algoritmo K-means. Cuanto menor sea la inercia, mejor ser√° el modelo.

2. M√©todo del Codo (Elbow Method): Este m√©todo te permite visualmente identificar un buen n√∫mero de clusters. Grafica el valor de la inercia en funci√≥n del n√∫mero de clusters y busca el punto donde la disminuci√≥n en la inercia comienza a disminuir de manera menos pronunciada, formando un "codo". Este n√∫mero de clusters es una buena elecci√≥n.

3. √çndice de Silueta (Silhouette Score): Esta m√©trica cuantifica qu√© tan cerca est√°n los puntos de un cluster de los puntos de los clusters vecinos. Un valor alto indica que los puntos est√°n bien agrupados y est√°n lejos de los otros clusters. Es una medida de cohesi√≥n y separaci√≥n entre clusters. Un valor entre -1 y 1, donde valores m√°s cercanos a 1 son mejores.

4. Distribuci√≥n de Tama√±os de Cluster: Es importante entender si los clusters tienen tama√±os similares o si algunos clusters son mucho m√°s grandes que otros. Puedes visualizar la distribuci√≥n de tama√±os de cluster para entender mejor la calidad de la agrupaci√≥n.

5. Validaci√≥n Externa (si tienes etiquetas verdaderas): Si tienes etiquetas verdaderas para tus datos, puedes usar m√©tricas como el √çndice de Rand Ajustado (Adjusted Rand Index) o el √çndice de Jaccard para comparar las etiquetas reales con las asignaciones de clusters.

6. Visualizaci√≥n de Clusters: Puedes visualizar los clusters en un espacio de baja dimensi√≥n (por ejemplo, usando PCA o t-SNE) para entender si los clusters formados tienen sentido intuitivo y se alinean con la estructura de los datos.

Recuerda que no hay una √∫nica m√©trica que sea definitiva en todos los casos. Es importante utilizar m√∫ltiples m√©tricas y t√©cnicas de evaluaci√≥n para tener una comprensi√≥n completa del rendimiento del modelo K-means. Adem√°s, es posible que algunas m√©tricas sean m√°s apropiadas dependiendo del contexto espec√≠fico de tu problema y de los datos con los que est√©s trabajando.
